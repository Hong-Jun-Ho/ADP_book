각 선수는 id로 구분되어 있으며, 'g' 칼럼은 각 선수가 해당 연도에 출전한 게임 수를 나타냄

```python
from rpy2.robjects.packages import importr   # r의 패키지를 불러옴
plyr = importr("plyr")    # baseball 데이터는 plyr의 내장 데이터이므로 plyr을 불러옴
r.data('baseball')        # r의 샘플데이터를 불러옴
baseball = r['baseball']
print(baseball.head())

baseball['avgG'] = baseball.groupby('id')['g'].transform('sum') / baseball.groupby('id')['year'].transform('count')
# transform 함수로 각 id 별 모든 연도의 출전횟수 합과 출전한 연도의 수를 구하여 나눔
# baseball.groupby('id')['g'].transform('sum')는 baseball 데이터를 id에 대해 그룹화한 후 칼럼 'g'에 대하여 sum을 가져오는 것을 의미

baseball.loc[baseball['id']=='aaronha01'].head() # 결과 확인
```

Cars93 데이터를 Price가 12 미만이면 low, 23 미만이면 middle, 23 이상이면 high 값을 가지는 변수 Pr_level을 생성한 후, Model, Price, Pr_level만 출력하라

```python
@make_symbolic
def np_where(bools, val_if_true, val_if_false):
    return np.where(bools, val_if_true, val_if_false)

Cars93 >> mutate(Pr_level = np_where(X.Price < 12, 'low', np_where(X.Price < 23, 'middle', 'high'))) >> select(X.Model, X.Price, X.Pr_level)
```

위의 코드에서는 dfply로 문제를 해결하기 위해 @make_symbolic을 사용하여 np.where(조건문, 'yes', 'no') 함수를 불러왔다.

@make_symbolic은 dfply에서 사용자 정의 함수를 생성하는 데 사용하며, np_where는 통용되어 선언되는 함수 중 하나이다.

하지만 이는 쉽게 생각해내기 어려우므로 Pandas DataFrame에서 해결할 수 있는 코드를 아래에 제공한다.

```python
Cars93['Pr_level'] = Cars93.apply(lambda X: 'low' if X.Price < 12 else ('middle' if X.Price < 23 else 'high'), axis=1)
Cars93[['Model', 'Price', 'Pr_level']]
```

Weight 변수들의 값이 중앙값 미만인 경우 "low", 이상인 경우 "high"를 갖는 변수 Weight_lv를 생성하라<br>
그리고 Price를 기준으로 데이터를 오름차순 정렬하라

- outer_join(b, by='x1') / full_join(b, by='x1')
- left_join(b, by='x1')
- right_join(b, by='x1')

> a >> inner_join(b, by='x1')

<br>

<b>[예제]</b> dplyr.join => dfply.join

카페에서 판매하는 메뉴의 code, name을 담은 데이터 'NAME'과 메뉴의 code, price를 담은 데이터 'PRICE'를 생성하고,
각 메뉴의 고유 코드를 의미하는 code 변수를 기준으로 left_join, right_join, full_join을 수행하라

```python
NAME = DataFrame({'code': ['A1', 'A2', 'A3'],
                  'name': ['coffee', 'tea', 'cookie']})

PRICE = DataFrame({'code': ['A1', 'A2', 'A4'],
                   'price': [2000, 2500, 4000]})

# inner_join
NAME >> inner_join(PRICE, by='code')
```

```python
# outer_join
NAME >> outer_join(PRICE, by='code')
```

```python
# left_join
NAME >> left_join(PRICE, by='code')
```

```python
# right_join
NAME >> right_join(PRICE, by='code')
```

#### bind_rows() : 데이터의 행들을 이어붙여 결합하는 함수
> bind_rows(other, join='inner'/'outer', ignore_index=False)

<br>

<b>[예제]</b> dplyr.bind_rows => dfply.bind_rows

앞서 생성한 'NAME'과 'PRICE'에 대해 행 기준 결합을 수행하라

```python
NAME >> bind_rows(PRICE,  join='outer')
```

<b>[예제]</b> dplyr.bind_rows => dfply.bind_rows

카페 메뉴의 코드와 이름을 가진 세 개의 데이터를 생성하고 이들을 행으로 결합하며, 어떤 데이터로부터 결합한 것인지 id도 나타내라

```python
A = DataFrame({'code': ['A1', 'A2'],
               'name': ['coffee', 'tea']})
B = DataFrame({'code': ['A3', 'A4'],
               'name': ['cookie', 'cake']})
C = DataFrame({'code': ['A5'],
               'name': ['juice']})

A >> mutate(id = 'A') >> bind_rows(B >> mutate(id = 'B'),  join='outer') >> bind_rows(C >> mutate(id = 'C'), join='outer')
```

#### bind_cols() : 데이터의 열을 이어붙여 결합하는 함수
> bind_cols(other, join='inner'/'outer', ignore_index=False)

<br>

<b>[예제]</b> dplyr.bind_cols => dfply.bind_cols

다섯 명의 학생의 국어점수와 수학점수가 저장된 데이터를 각각 생성하고 두 데이터를 열로 결합하라

```python
kor = DataFrame({'id': ['A1', 'A2', 'A3', 'A4', 'A5'],
                 'kor': [80, 75, 56, 69, 90]})

math = DataFrame({'id': ['A1', 'A2', 'A3', 'A4', 'A5'],
                  'math': [91, 68, 68, 73, 98]})

kor >> bind_cols(math,  join='outer') >> select(columns_from('kor'))
```
